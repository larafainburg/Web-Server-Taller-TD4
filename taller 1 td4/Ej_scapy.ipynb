### Creemos un paquete

```python
ip_destino = "8.8.8.8"  # IP de Google

puerto_destino = 53  # Puerto DNS

puerto_origen = 1234  # Puerto origen cualquiera

# Creamos el paquete
udp_packet = IP(dst=ip_destino) / UDP(sport=puerto_origen, dport=puerto_destino) / Raw(load="hola")

udp_packet.show()  # Inspeccionamos el paquete antes de mandarlo

# Mandamos el paquete
send(udp_packet)
```
